```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.model_selection import train_test_split
```

```python
def split_scaler(indep,dep):
    x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=0.25,random_state=0)
    sc=StandardScaler()
    x_train=sc.fit_transform(x_train)
    x_test=sc.transform(x_test)
    return x_train,x_test,y_train,y_test
```

```python
def rfefeature(indep,dep,n):
    rfelist=[]
    log_reg=LogisticRegression()
    rf=RandomForestClassifier()
    des=DecisionTreeClassifier()
    svml=SVC(kernel='linear')
    #svmnl=SVC(kernel='rbf')
    rfemodellist=[log_reg,rf,des,svml]
    for i in rfemodellist:
        print (i)
        rfe=RFE(i,n)
        rfefit=rfe.fit(indep,dep)
        logrfefeature=rfefit.transform(indep)
        rfelist.append(logrfefeature)
    return rfelist
    
```

```python
def cm_prediction(classifier,x_test):
    y_pred=classifier.predict(x_test)
    from sklearn.metrics import confusion_matrix
    from sklearn.metrics import classification_report
    from sklearn.metrics import accuracy_score
    cm=confusion_matrix(y_test,y_pred)
    report=classification_report(y_test,y_pred)
    accuracy=accuracy_score(y_test,y_pred)
    return cm,accuracy,report,x_test,y_test,classifier
    
```

```python
def logistic(x_train,y_train,x_test):
    from sklearn.linear_model import LogisticRegression
    classifier =LogisticRegression()
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
def svm_linear(x_train,y_train,x_test):
    from sklearn.svm import SVC
    classifier=SVC(kernel='linear')
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
def svmnl(x_train,y_train,x_test):
    from sklearn.svm import SVC
    classifier=SVC(kernel='rbf')
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
def des(x_train,y_train,x_test):
    from sklearn.tree import DecisionTreeClassifier
    classifier=DecisionTreeClassifier()
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
def rf(x_train,y_train,x_test):
    from sklearn.ensemble import RandomForestClassifier
    classifier=RandomForestClassifier()
    classifier.fit(x_train,y_train)
    cm,accuracy,report,x_test,y_test,classifier=cm_prediction(classifier,x_test)
    return cm,accuracy,report,x_test,y_test,classifier
```

```python
def rfe_feature(acclog,accsvml,accdecision,accrandom):
    rfedataframe=pd.DataFrame(index=['Logistic','SVML','Decision','Random'],columns=['Logistic','SVML','Decision','Random'])
    for number,index in enumerate(rfedataframe.index):
        rfedataframe['Logistic'][index]=acclog[number]
        rfedataframe['SVML'][index]=accsvml[number]
        #dataframe['SVMNL'][index]=accsvmnl[number]
        #dataframe['Naive'][index]=accnaive[number]
        #dataframe['KNN'][index]=accknn[number]
        rfedataframe['Decision'][index]=accdecision[number]
        rfedataframe['Random'][index]=accrandom[number]
        return rfedataframe                        
    
```

```python
df1=pd.read_csv("CKD.csv")
df2=df1

df2=pd.get_dummies(df2,drop_first=True)
indep=df2.drop('classification_yes',axis=1)
dep=df2['classification_yes']
```

```python
rfelist=rfefeature(indep,dep,6)

acclog=[]
accsvml=[]
accdecision=[]
accrandom=[]


for i in rfelist:
    
    x_train,x_test,y_train,y_test=split_scaler(i,dep)
    
    cm,accuracy,report,x_test,y_test,classifier=logistic(x_train,y_train,x_test)
    acclog.append(accuracy)

    cm,accuracy,report,x_test,y_test,classifier=svm_linear(x_train,y_train,x_test)
    accsvml.append(accuracy)


    cm,accuracy,report,x_test,y_test,classifier=des(x_train,y_train,x_test)
    accdecision.append(accuracy)

    cm,accuracy,report,x_test,y_test,classifier=rf(x_train,y_train,x_test)
    accrandom.append(accuracy)
    
```
